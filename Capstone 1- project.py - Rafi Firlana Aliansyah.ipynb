```python
# ---------------- DATA AWAL----------------

procurements = [{
    "confidential": "Confidential",
    "item": "Barang",
    "divisi": "Finance",
    "detail_project": "survey kepuasan pelanggan",
    "harga": 24000000,
    "status": "Belum Dibayar"
}]


next_id = 1


def inisialisasi_id():
    global next_id

    highest_id = 0

    for index, p in enumerate(procurements, start=1):
        p["no"] = index
        highest_id = index

    next_id = highest_id + 1

# ---------------- MENU 1 ----------------

def tampilkan_procurement():
    if not procurements:
        print("Belum ada data procurement.\n")
        return

    print("\nDaftar Procurement:")
    for p in procurements:
        print(
            f"- No: {p['no']} | "
            f"Confidential: {p['confidential']} | "
            f"Item: {p['item']} | "
            f"Divisi: {p['divisi']} | "
            f"Detail Project: {p['detail_project']} | "
            f"Harga: Rp {p['harga']:,} | "
            f"Status: {p['status']}"
        )
    print()



# ---------------- MENU 2 ----------------
def pilih_dari_menu(judul, opsi):
    print(f"\nPilih {judul}:")
    for i, o in enumerate(opsi, start=1):
        print(f"{i}. {o}")

    try:
        pilihan = int(input("Masukkan nomor pilihan: "))
        if 1 <= pilihan <= len(opsi):
            return opsi[pilihan - 1]
        else:
            print("❌ Pilihan tidak valid")
            return None
    except ValueError:
        print("❌ Input harus berupa angka")
        return None


def tambah_procurement():
    global next_id
    
    
 # ✅ AMBIL ID SEKARANG
    no = next_id
    next_id += 1
    
# ✅ Tampilkan data terlebih dahulu (seperti Menu 1)
    tampilkan_procurement()

    confidential_opsi = [
        "Confidential",
        "Not Confidential"
    ]

    item_opsi = [
        "Barang",
        "Jasa / Service",
        "Event",
        "Vehicle Logistics",
        "Survey",
        "Catering",
        "Tour & Travel",
        "Training",
        "Item",
        "Import"
    ]

    divisi_opsi = [
        "Marketing",
        "Sales",
        "Operation",
        "Finance",
        "HRD"
    ]

    
# ---------- Confidential ----------
    while True:
        confidential = pilih_dari_menu("Status Confidential", confidential_opsi)
        if confidential:
            break
        print("⚠️ Pilihan tidak valid, silakan masukkan ulang.")


# ---------- Jenis Item ----------
    
    while True:
        item = pilih_dari_menu("Jenis Item", item_opsi)
        if item:
            break
        print("⚠️ Pilihan tidak valid, silakan masukkan ulang.")


# ---------- Divisi ----------

    while True:
        divisi = pilih_dari_menu("Divisi", divisi_opsi)
        if divisi:
            break
        print("⚠️ Pilihan tidak valid, silakan masukkan ulang.")

    
# ---------- Detail Project ----------

    detail_project = input("Masukkan detail project: ")

# ---------- Harga ----------

   
    while True:
        try:
            harga = int(input("Masukkan harga (minimal 4 angka): "))
            if harga < 1000:
                print("❌ Harga minimal 4 angka (≥ 1000)")
            else:
                break
        except ValueError:
            print("❌ Harga harus berupa angka")


    procurements.append({
        "no": no,
        "confidential": confidential,
        "item": item,
        "divisi": divisi,
        "detail_project": detail_project,
        "harga": harga,
        "status": "Belum Dibayar"
    })

    print(f"✅ Data ditambahkan (ID: {no})\n")


# ---------------- MENU 3 ----------------
def hapus_procurement():
    # Rule 1: jika database kosong
    if not procurements:
        print("Belum ada data yang bisa dihapus\n")
        return  # kembali ke menu utama

    
 # ✅ Tampilkan data terlebih dahulu (seperti Menu 1)
    tampilkan_procurement()

    while True:
        try:
            no_hapus = int(input("Masukkan nomor procurement yang ingin dihapus: "))
        except ValueError:
            print("❌ Nomor procurement harus berupa angka")
            continue

        # Cari data berdasarkan nomor
        for i, p in enumerate(procurements):
            if p["no"] == no_hapus:
                del procurements[i]
                print("✅ Procurement berhasil dihapus\n")
                return  # kembali ke menu utama

        # Rule 2: jika tidak ditemukan
        print("❌ Nomor procurement tidak ditemukan")





# ---------------- MENU 4 ----------------
# Update Data


def update_procurement():
    # Rule 1: jika data kosong
    if not procurements:
        print("Belum ada data yang bisa diupdate\n")
        return

    tampilkan_procurement()

    # 🔁 LOOP INPUT NOMOR (RULE 2)
    while True:
        try:
            no_update = int(input("Masukkan nomor procurement yang ingin diupdate: "))
        except ValueError:
            print("❌ Nomor harus berupa angka\n")
            continue

        for p in procurements:
            if p["no"] == no_update:
                break
        else:
            # ✅ RULE 2
            print("❌ Nomor procurement tidak ditemukan, silakan masukkan ulang\n")
            continue

        # keluar dari loop input nomor
        break

    # 🔧 MENU UPDATE
    while True:
        print("\nData saat ini:")
        print(
            f"Confidential: {p['confidential']} | "
            f"Item: {p['item']} | "
            f"Divisi: {p['divisi']} | "
            f"Detail Project: {p['detail_project']} | "
            f"Harga: Rp {p['harga']:,} | "
            f"Status: {p['status']}"
        )

        print("\nPilih data yang ingin diupdate:")
        print("1. Confidential")
        print("2. Jenis Item")
        print("3. Divisi")
        print("4. Detail Project")
        print("5. Harga")
        print("6. Status")
        print("0. Selesai Update")

        pilihan = input("Masukkan pilihan: ")

        if pilihan == "1":
            hasil = pilih_dari_menu("Status Confidential", ["Confidential", "Not Confidential"])
            if hasil:
                p["confidential"] = hasil

        elif pilihan == "2":
            hasil = pilih_dari_menu(
                "Jenis Item",
                [
                    "Barang", "Jasa / Service", "Event", "Vehicle Logistics",
                    "Survey", "Catering", "Tour & Travel", "Training", "Item", "Import"
                ]
            )
            if hasil:
                p["item"] = hasil

        elif pilihan == "3":
            hasil = pilih_dari_menu("Divisi", ["Marketing", "Sales", "Operation", "Finance", "HRD"])
            if hasil:
                p["divisi"] = hasil

        elif pilihan == "4":
            p["detail_project"] = input("Masukkan detail project baru: ")

        elif pilihan == "5":
            while True:
                try:
                    harga_baru = int(input("Masukkan harga baru (≥1000): "))
                    if harga_baru >= 1000:
                        p["harga"] = harga_baru
                        print("✅ Harga berhasil diperbarui")
                        break
                except ValueError:
                    print("❌ Harga harus berupa angka")

        elif pilihan == "6":
            hasil = pilih_dari_menu("Status Pembayaran", ["Belum Dibayar", "Sudah Dibayar"])
            if hasil:
                p["status"] = hasil

        elif pilihan == "0":
            print("✅ Update selesai\n")
            return

        else:
            print("❌ Pilihan tidak valid")

    


# ---------------- MENU 5 ----------------

!pip install matplotlib
import matplotlib.pyplot as plt
def kesimpulan_procurement():
    if not procurements:
        print("Belum ada data.\n")
        return

    total = sum(p["harga"] for p in procurements)
    sudah_dibayar = sum(
        p["harga"] for p in procurements if p["status"] == "Sudah Dibayar"
    )
    sisa = total - sudah_dibayar

    # ---------- OUTPUT TEKS ----------
    print("\n📊 Kesimpulan Procurement:")
    print(f"Total Procurement      : {len(procurements)}")
    print(f"Total Budget           : Rp {total:,}")
    print(f"Total Sudah Dibayar    : Rp {sudah_dibayar:,}")
    print(f"Sisa Budget            : Rp {sisa:,}\n")

    # ---------- GRAFIK ----------
    labels = ["Total Budget", "Sudah Dibayar", "Sisa Budget"]
    values = [total, sudah_dibayar, sisa]
    colors = ["blue", "green", "red"]

    plt.figure(figsize=(8, 5))
    plt.bar(labels, values, color=colors)

    plt.title("Visualisasi Budget Procurement Toyota")
    plt.ylabel("Nilai (Rupiah)")
    plt.xlabel("Kategori")

    for i, v in enumerate(values):
        plt.text(i, v, f"Rp {v:,}", ha="center", va="bottom")

    plt.tight_layout()
    plt.show()



# ---------------- MAIN LOOP ----------------
def chatbot():
    while True:
        print("""
Selamat datang di Procurement Management System Toyota

Masukkan angka menu yang ingin dijalankan:

1. Menampilkan daftar Procurement yang sudah di input
2. Menambah list Procurement
3. Menghapus list procurement
4. Update Data procurement
5. Melihat kesimpulan pembayaran procurement
6. Exit Program
""")

        menu = input("Pilih menu (1-6): ")

        if menu == "1":
            tampilkan_procurement()
        elif menu == "2":
            tambah_procurement()
        elif menu == "3":
            hapus_procurement()
        elif menu == "4":
            update_procurement()
        elif menu == "5":
            kesimpulan_procurement()
        elif menu == "6":
            print("Terima kasih, program selesai.")
            break
        else:
            print("⚠️ Menu tidak valid!\n")


# ---------------- RUN PROGRAM ----------------

inisialisasi_id()

if __name__ == "__main__":
    chatbot()
```
